Find the eigenvalues and eigenvectors of a matrix and plot the eigenvectors.

```python


import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a matrix
A = np.array(
    [[1, 2],
     [3, 4]])
print("Matrix A :", A)

# Step 2: Find the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 3: Plot the eigenvectors
origin = [0, 0]

fig, ax = plt.subplots()
ax.quiver(*origin, *eigenvectors[:,0], angles='xy', scale_units='xy', scale=1, color='r', label='Eigenvector 1')
ax.quiver(*origin, *eigenvectors[:,1], angles='xy', scale_units='xy', scale=1, color='b', label='Eigenvector 2')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
ax.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Eigenvectors of the Matrix A')
plt.show()

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)
```
