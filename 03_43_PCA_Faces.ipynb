PCA Faces

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.io

# Check if the file exists in the Google Colab current directory
if not os.path.exists('allFaces.mat'):
    # If the file doesn't exist, download it
    !wget https://github.com/kursatkara/databook_python/raw/master/DATA/allFaces.mat
else:
    print("allFaces.mat file already exists.")

# Load the mat file
mat_contents = scipy.io.loadmat('allFaces.mat')
```

```python

plt.rcParams['figure.figsize'] = [10, 10]
plt.rcParams.update({'font.size': 18})

# Extract the necessary data
faces = mat_contents['faces']

m = int(np.squeeze(mat_contents['m']))
n = int(np.squeeze(mat_contents['n']))

nfaces = np.ndarray.flatten(mat_contents['nfaces'])

allPersons = np.zeros((n*6,m*6))
count = 0

for j in range(6):
    for k in range(6):
        allPersons[j*n : (j+1)*n, k*m : (k+1)*m] = np.reshape(faces[:,np.sum(nfaces[:count])],(m,n)).T
        count += 1

img = plt.imshow(allPersons)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```
